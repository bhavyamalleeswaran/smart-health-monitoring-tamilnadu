```python
df = pd.read_csv('../data/tamil_nadu_waterborne_disease_data.csv')
print(f'✅ Dataset loaded!')
print(f'📊 Rows: {df.shape[0]}')
print(f'📋 Columns: {df.shape[1]}')
print(f'📅 Date Range: {df["date"].min()} to {df["date"].max()}')
print(f'🗺️ Districts: {df["district"].nunique()}')
```

```python
missing = df.isnull().sum()
print('Missing Values:')
print(missing)
print(f'\n✅ Total missing values: {missing.sum()}')
```

```python
outbreak_count = df['outbreak_alert'].value_counts()
print('Outbreak Alert Distribution:')
print(outbreak_count)
print(f'\nOutbreak Rate: {(outbreak_count.sum())}  total records')
```

```python
print('Missing Values:')
print(df.isnull().sum())
print(f'\n✅ Total missing values: {df.isnull().sum().sum()}')
```

```python
district_summary = df.groupby('district')['total_disease_cases'].sum()
district_summary = district_summary.sort_values(ascending=False)
print('Total Cases by District:')
print(district_summary)
```
